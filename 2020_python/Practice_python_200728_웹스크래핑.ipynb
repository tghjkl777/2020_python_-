```python
from bs4 import BeautifulSoup #파서 역할

#테스트용 html 코드
html="""<html>
<head>
<title>웹스크래핑</title>
</head>
            <body>
                <div>
                    <span>
<a href=http://www.naver.com>naver</a>
<a href=http://www.google.com>google</a>
                    </span>
                </div>
                <div class='hi', id='lower'>
                <p>aaaaaaaaaaa</p>
                <p>bbbbbbbbbbb</p>
                </body>
            </html>"""

soup=BeautifulSoup(html,'lxml')
soup
```

```python
print(soup.prettify())
```

```python
soup.find('a') #태그에 해당하는 첫번째 줄 리턴함
```

```python
soup.
```

```python
soup.find('a').get_text() #태그 안에 있는 내용만 얻고 싶다.
```

```python
soup.find_all('p') #일치하는 태그의 모든 내용 본다,.
```

```python
soup.find('div',id='lower',class_='hi') # 클래스는  파이썬의 예약어이기 때문에 언더바를 붙인다. 
```

```python
arrts={'id':'lower','class':'hi'}
print(soup.find('dic',attrs=arrts))
```

```python
#일치하는 태그 둘의 모든 내용들만 보기
site_name=soup.find_all('a')
for site in site_name :
    print(site.get_text())  
```

```python
html2="""<!DOCTYPE html>
<html>
<head>
<title>작퓸과 작가 모음</title>
</head>
<body>
<h1>책 정보</h1>
  <p id="book_title">토지</p>
  <p id="author">박경리</p>
  
  <p id="book_title">태백산맥</p>
  <p id="author">조정래</p>
  
    <p id="book_title">감옥으로부터의 사색</p>
  <p id="author">신영복</p>
</body>
</html>
"""

soup2=BeautifulSoup(html2,'lxml')
soup2
```

```python
soup2.find_all('p')
```

```python
#책 제목, 작가 따로 가져오기 
soup2.find('p',{'id':'book_title'})
```

```python
a=soup2.find_all('p',id='book_title')
```

```python
b=soup2.find_all('p',id='author')
```

```python
for l, u in list(zip(a,b)):#책 제목, 작가 따로 가져오기 
    print(l.get_text(),u.get_text())
```
