```python
def hello(count):
    if count==0 :
        return
    print("je",count)
    count-=1
    hello(count)
hello(5)  #재귀 함수 사용할때 한다. 재귀함수 코드 굉장히 간결하지만 엄청 선호하지는 않는다. 

def factorial(n):
    if n==1 :
        return 1
    return n*factorial(n-1)

print(factorial(5))
```

### 객체와 클래스

함수는 기능 하나라고 보고 객체는 몽땅 모인것
추상화 : 객체를 보고 어떤 기능과 속성이 필요한느지 생각하는것
클래스는 대문자로 시작
함수는 소문자로 시작

```python
class Person:
    #생성자
    def __init__(self):
        self.name='carami'
    
    #amethon
    def hello(self):
        print(self.name,"dkssud")
        
carami=Person()
carami.hello()

isinstance(carami,Person) #객체의 인스터스인지 확인하는 함수
```

1. self
self는 객체의 인스턴스 그 자체를 의미합니다.
대부분 객체지향 언어는 이걸 메소드에 안 보이게 전달하지만 파이썬에서 클래스의 메소드를 정의할 때는 self를 꼭 명시해하고
그 메소드를 불러올 때 self는 자동으로 전달됩니다.

2. __init__
__init__은 파이썬에서 쓰이는 생성자입니다.

위의 코드에서 A()는 생성자 __init__에 어떤 파라미터도 넘기지 않고, 그 결과로 A타입의 객체를 생성해 이를 반환받습니다.

A(24, 'Hello'))와 같이 쓰면 파라미터 2개를 받는 생성자가 필요한데,
현재 __init__은 그 어떤 파라미터도 받지 않으니 exception이 발생합니다.

```python
class Person :
    def __init__ (self, name):
        self.name=name
    def hello(self):
        print(self.name)
        
a=Person("park")
a.hello()
```

```python
class Car :
    instance_count=0
    
    def __init__ (self, size, color):
        self.size=size
        self.color=color
        Car.instance_count+=1
        print("자동차 수 {}".format(Car.instance_count))
        
    def move (self):
        print("자동차 {} 와 {} 가 움직임".format(self.size, self.color))

              
car1=Car(100,"black")
car2=Car(200,'white')

print("num instance {}".format(Car.instance_count))

print(car1.instance_count, car2.instance_count)


car1.move()
car2.move()
```
