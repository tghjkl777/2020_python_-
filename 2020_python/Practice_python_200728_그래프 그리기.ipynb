### matplolit 산점도

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']='False'  #한글 쓸수 있게
%matplotlib inline
```

```python
height=[165,177,160,180,185,155,172]
weight=[62,67,55,74,90,40,64]

plt.scatter(height,weight) # 산점도그리기
plt.xlabel("height")
plt.ylabel("weight")
plt.title('h&w')
plt.grid()
```

```python
plt.scatter(height,weight,s=500,c='r') #동일한 마커의 컬러 앤 크기 지정
```

```python
size=100*np.arange(1,8) 
colors=['r','g','b','c','m','k','y']

plt.scatter(height,weight,s=size,c=colors) #마커 크기 색깔 다 다르게
```

```python
city=['서울','인천','대전','대구','울산','부산','광주']

lat=[37.56,37.45,36.35,35.87,35.87,35.53,35.16]

lon=[126.97,126.70,127.35,128.60,129.31,129.07,126.85]

pop_den=[16154,2751,2389,2790,1099,4454,2995]

size=np.array(pop_den)
colors=['r','g','b','c','m','k','y']

plt.scatter(lon,lat,s=size,c=colors,alpha=0.5)
plt.xlabel("lon")
plt.ylabel("lat")
plt.title('Density')

for x,y,name in zip(lon,lat,city) : #위도 경도에 맞는 도시 이름 출력
    plt.text(x,y,name)
```

### 막대그래프

```python
m=['m_01','m_02','m_03','m_04']
b=[27,35,40,33]
a=[30,38,42,37]

n_data=len(m)
index=np.arange(n_data)
plt.bar(index,b) 막대그프래 그리기
plt.show()
```

```python
plt.bar(index,b,tick_label=m) #x축 라벨 붙이기.
```

```python
colors=['r','g','b','m']
plt.bar(index,b,tick_label=m,color=colors) #색깔 추가
```

```python
plt.bar(index,b,tick_label=m,color=colors,width=0.6) #폭변경
```

```python
plt.barh(index,b,tick_label=m,color=colors)  #가로 그래프로 그리기
```

### 파이 그래프

```python
fruit=['apple','banan','strawberry','orange','grape']
result=[7,6,3,2,2]

plt.pie(result)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(result)
plt.show()
```

```python
plt.pie(result,labels=fruit)#해당 이름 보여주기
plt.show()
```

#### Data_Frame 이용한 그래프

```python
df_rain=pd.read_csv("sea_rain1.csv",index_col='연도')
df_rain
```

```python
df_rain.plot()
plt.show()  #판다스로 그리기
```

```python
rain_plot=df_rain.plot(grid=True, style=['r-*','g-o','b:*','m-p'])
rain_plot.set_xlabel("연도")
rain_plot.set_ylabel("강수량")
rain_plot.set_title("연간 강수량")
plt.show()
```

```python
year=[2006,2008,2010,2012,2014,2016]
area=[26,28,28,31,34,33]
table={'연도':year ,
       '주거면적' :area}
df_area=pd.DataFrame(table,columns=['연도','주거면적'])
df_area
```

```python
df_area.plot(x= '연도' ,y='주거면적', grid=True, title='연도별 주거면적')
plt.show()
```
