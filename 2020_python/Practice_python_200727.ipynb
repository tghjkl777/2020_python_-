```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv("train.csv")
```

```python
df_train
```

```python
df_train.head()
```

```python
#group by  -그룹핑 해야할때 어떤 기준으로 어떻게 할것이냐. 어떤 컬럼을 기준으로 그루핑 할것이다. 

pclass_group=df_train.groupby('Pclass')
pclass_group.groups
```

```python
pclass_group.mean()['Survived']
```

```python
df_train.set_index('Age').groupby(level=0).mean()['Survived']
```

```python
df_train.groupby(['Pclass','Sex']).mean()
```

```python
df_train.groupby(['Pclass','Sex']).mean()['Survived']
```

```python
df_train.groupby(['Pclass','Age']).aggregate(np.sum,np.mean()) 
```

```python
df_train.groupby('Pclass').mean() #평균값을 보여준다.
```

```python
df_train.groupby('Pclass').transform(np.mean) #평균값을 어쨋듯보여 주는데 각각 보여준다.
```

```python
data_list= [ 
    {'yyyy-mm-dd':'2003-04-18'},
       {'yyyy-mm-dd':'2010-05-12'}, 
        {'yyyy-mm-dd':'2020-07-27'}
]
df=pd.DataFrame(data_list , columns=['yyyy-mm-dd'])
df
```

```python
# 년도 값을 구해서 년도 값을 새로운 컬럼에다가 넣어보고 싶다. apply함수를 이용해야 한다. 
def get_year(colum) :
    return colum.split('-')[0]

df['year']=df['yyyy-mm-dd'].apply(get_year)
df
```
