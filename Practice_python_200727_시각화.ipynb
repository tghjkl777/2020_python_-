```python
#하나의 그래프 창을 하위 그래프 영억으로 나누기 위한것



x=np.arange(0,10,0.1)
y1=0.3*(x-5)**2+1
y2=-1.5*x+3
y3=np.sin(x)**2
y4=10*np.exp(-x)+1

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.plot(x,y2)
plt.cla() # 전에 지정된 3번 그래프 지우기.

plt.subplot(2,2,4)  ##비우고 싶으면 지운다음에 다시 쓰기
plt.plot(x,y4)
```

## 그래프 꾸미기 
-  plot의 fmt 옵션 : 선색깔 ,선 스타일, 마커 지정가능.

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1,x,y2,x,y3,x,y4) #컬러지정 안해도 알아서 다양한 색으로 보여줌
```

```python
plt.plot(x,y1,'m',x,y2,'y',x,y3,'k',x,y4,'c') #색깔 바꾸기
```

```python
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.') #선스타일 다르게
```

```python
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d') #마커 지정
```

```python
plt.plot(x,y1,'>--r',x,y2,'^k:',x,y3,'s-g',x,y4,'-.cX')  #혼합 사용가능
```

```python
x=np.arange(-4.5,5,0.5)
y=2*x**3

plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")  #축 이름 쓰기
plt.title("Graph title") #그래프 제목쓰기
plt.grid() # 격자 추가
plt.show()
```

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.legend(['data1','data2','data3','data4']) #범례 써주기
```

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3

plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.legend(['data1','data2','data3','data4'],loc='lower right')  #범례 위치 지정.
```
